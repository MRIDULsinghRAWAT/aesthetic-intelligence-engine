```python
# ============ COMPLETE SETUP - RAM OPTIMIZED ============

from google.colab import drive
drive.mount('/content/drive')

import zipfile, os
import pandas as pd
import numpy as np
import cv2
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Extract
zip_path = '/content/drive/MyDrive/aesthetic-engine-data/archive (4).zip'
extract_path = '/content/dataset/'

if not os.path.exists(extract_path + 'BIQ2021.csv'):
    print("Extracting...")
    with zipfile.ZipFile(zip_path, 'r') as z:
        z.extractall(extract_path)
else:
    print("Already extracted!")

# Load CSV - SIRF 2000 images
df = pd.read_csv('/content/dataset/BIQ2021.csv')
df['label'] = (df['MOS'] > 0.5).astype(int)
df = df.sample(2000, random_state=42)  # ← SIRF 2000
print(f"Using: {len(df)} | Good: {df['label'].sum()} | Bad: {len(df)-df['label'].sum()}")

# Train/Val split
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Load Images - SMALLER SIZE
def load_images_batch(dataframe, img_dir, img_size=(128, 128)):  # ← 128 not 224
    images, labels = [], []
    for _, row in dataframe.iterrows():
        path = img_dir + row['Images']
        img = cv2.imread(path)
        if img is not None:
            img = cv2.resize(img, img_size) / 255.0
            images.append(img)
            labels.append(row['label'])
    return np.array(images, dtype=np.float32), np.array(labels)

print("Loading images... (1-2 min)")
X_train, y_train = load_images_batch(train_df, '/content/dataset/')
X_val, y_val     = load_images_batch(val_df,   '/content/dataset/')

print(f"X_train: {X_train.shape}")
print(f"X_val:   {X_val.shape}")
print("ALL DONE!")
```

```python
# ============ MODEL + TRAINING ============

from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, Model

# MobileNetV2 base
base = MobileNetV2(input_shape=(128, 128, 3),
                   include_top=False,
                   weights='imagenet')
base.trainable = False

# Custom head
x = base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
out = layers.Dense(1, activation='sigmoid')(x)

model = Model(inputs=base.input, outputs=out)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print(f"Model params: {model.count_params():,}")

# Train
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_val, y_val),
    verbose=1
)

# Save
model.save('/content/aesthetic_model.h5')
print("Model saved!")
```

```python
# ============ TFLITE CONVERSION ============
import os

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
with open('/content/aesthetic_model.tflite', 'wb') as f:
    f.write(tflite_model)

converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_quant = converter.convert()
with open('/content/aesthetic_model_quantized.tflite', 'wb') as f:
    f.write(tflite_quant)

h5_size    = os.path.getsize('/content/aesthetic_model.h5') / 1e6
tflite_size = os.path.getsize('/content/aesthetic_model.tflite') / 1e6
quant_size  = os.path.getsize('/content/aesthetic_model_quantized.tflite') / 1e6

print(f"Original (.h5):   {h5_size:.2f} MB")
print(f"TFLite:           {tflite_size:.2f} MB")
print(f"TFLite Quantized: {quant_size:.2f} MB")
```

```python
# ============ BENCHMARK ============
import time
import numpy as np

def benchmark_model(model_path, test_image, runs=100):
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    img = np.expand_dims(test_image, axis=0).astype(np.float32)

    # Warmup
    interpreter.set_tensor(input_details[0]['index'], img)
    interpreter.invoke()

    # Benchmark
    start = time.time()
    for _ in range(runs):
        interpreter.set_tensor(input_details[0]['index'], img)
        interpreter.invoke()
    avg_latency = (time.time() - start) / runs * 1000

    score = interpreter.get_tensor(output_details[0]['index'])[0][0]
    return avg_latency, score

test_img = X_val[0]

lat1, score1 = benchmark_model('/content/aesthetic_model.tflite', test_img)
lat2, score2 = benchmark_model('/content/aesthetic_model_quantized.tflite', test_img)

print(f"TFLite:           {lat1:.2f} ms | Score: {score1:.3f}")
print(f"TFLite Quantized: {lat2:.2f} ms | Score: {score2:.3f}")
```
